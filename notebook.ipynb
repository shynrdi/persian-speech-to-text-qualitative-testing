# گزارش طراحی و اجرای آزمون‌های گفتار به متن فارسی (Offline, GPU)

## 1. مقدمه

هدف این نوت‌بوک، طراحی و اجرای مجموعه‌ای از آزمون‌های **قابل تکرار، مرحله‌بندی‌شده و آفلاین** برای ارزیابی مدل‌های متن‌باز گفتار به متن (ASR) با پشتیبانی از زبان فارسی است.
در این آزمون‌ها، تمرکز اصلی بر **دقت تشخیص گفتار فارسی** و **کارایی زمانی (latency و RTF)** مدل‌ها در سناریوی اجرای آفلاین روی **GPU** قرار دارد.

با توجه به تنوع بالای مدل‌ها و ابزارهای اینفرنس، اجرای هم‌زمان تمامی حالت‌ها منجر به پیچیدگی و دشواری تحلیل نتایج می‌شود. بنابراین، در این نوت‌بوک از یک **رویکرد ترتیبی (Staged Benchmarking)** استفاده شده است که در آن هر مرحله به یک پرسش مشخص پاسخ می‌دهد و نتایج آن، مبنای تصمیم‌گیری برای مرحله‌ی بعدی قرار می‌گیرد.

---

## 2. داده‌ی آزمون (Test Set)

برای جلوگیری از سوگیری و تضمین قابلیت مقایسه، در تمامی مراحل آزمون از **یک مجموعه صوت ثابت** به همراه **متن مرجع (Ground Truth)** استفاده می‌شود.

ویژگی‌های مجموعه‌ی آزمون:

* شامل ۲۰ تا ۳۰ فایل صوتی
* زبان: فارسی
* تنوع گوینده (چند نفر)
* پوشش سناریوهای زیر:

  * جملات ساده و روزمره
  * اعداد، تاریخ و واحد پول
  * نیم‌فاصله و نشانه‌گذاری
  * جملات بلند و چندبخشی
  * وام‌واژه‌ها و اسامی خاص
  * مکث، تپق و گفتار محاوره‌ای
  * تعداد محدودی نمونه‌ی دارای نویز

این مجموعه در تمامی مراحل بدون تغییر استفاده می‌شود تا نتایج به‌صورت مستقیم قابل مقایسه باشند.

---

## 3. مدل‌های مورد ارزیابی

مدل‌های مورد بررسی، همگی **متن‌باز** بوده و پشتیبانی از زبان فارسی دارند. تمرکز اصلی بر خانواده‌ی Whisper و چند مدل چندزبانه‌ی جایگزین است.

گروه‌های اصلی مدل‌ها:

* مدل‌های چندزبانه Whisper در اندازه‌های مختلف (از tiny تا large)
* مدل‌های بهینه‌شده برای اینفرنس سریع Whisper
* مدل‌های چندزبانه‌ی غیر-Whisper (مانند SeamlessM4T و MMS ASR)

در هر فاز، تنها زیرمجموعه‌ای از این مدل‌ها انتخاب می‌شود تا تحلیل نتایج ساده و هدفمند باقی بماند.

---

## 4. ابزارها و تنظیمات اینفرنس

تمامی آزمون‌ها به‌صورت **آفلاین** و روی **GPU** اجرا می‌شوند.
مدل‌ها از پیش دانلود شده و بارگذاری آن‌ها بدون نیاز به اتصال اینترنت انجام می‌شود.

ابزارهای اصلی اینفرنس شامل:

* چارچوب استاندارد PyTorch و Transformers
* موتورهای بهینه‌شده‌ی اینفرنس برای Whisper
* استفاده از محاسبات FP16 و در صورت نیاز quantization

برای اندازه‌گیری دقیق زمان اجرا روی GPU، همگام‌سازی CUDA قبل و بعد از اینفرنس انجام می‌شود تا زمان واقعی پردازش محاسبه گردد.

---

## 5. رویکرد ترتیبی آزمون‌ها

آزمون‌ها به چند **فاز متوالی** تقسیم شده‌اند. هر فاز یک هدف مشخص دارد، از جمله:

* بررسی صحت عملکرد اولیه
* مقایسه دقت مدل‌ها در اندازه‌های مختلف
* مقایسه‌ی ابزارهای مختلف اینفرنس برای یک مدل ثابت
* بررسی تأثیر بهینه‌سازی‌ها و quantization
* مقایسه Whisper با مدل‌های غیر-Whisper
* تحلیل خطاهای خاص زبان فارسی

اجرای هر فاز تنها پس از تحلیل نتایج فاز قبلی انجام می‌شود.

---

## 6. معیارهای ارزیابی

برای هر فایل صوتی و هر پیکربندی مدل، معیارهای زیر ثبت می‌شود:

* متن خروجی مدل
* زمان کل اینفرنس (ثانیه)
* طول فایل صوتی (ثانیه)
* نسبت زمان واقعی (RTF)
* نرخ خطای نویسه (CER) و در صورت نیاز نرخ خطای واژه (WER)

این معیارها امکان تحلیل هم‌زمان **دقت** و **کارایی زمانی** را فراهم می‌کنند.

---

## 7. خروجی و کاربرد نتایج

نتایج هر فاز به‌صورت جدول‌بندی‌شده در نوت‌بوک نمایش داده می‌شود و خلاصه‌ای تحلیلی از مشاهدات هر مرحله ارائه می‌گردد.

خروجی نهایی این نوت‌بوک می‌تواند به‌عنوان:

* مبنای انتخاب مدل برای استفاده در محصول
* بخش ارزیابی تجربی در پایان‌نامه یا مقاله
* مرجع مقایسه برای توسعه‌های بعدی

مورد استفاده قرار گیرد.

## Dataset Index


```python
import pandas as pd, numpy as np, librosa, soundfile as sf
from pathlib import Path
from tqdm.auto import tqdm

audio_root = Path("./test_samples")
audio_paths = sorted(audio_root.rglob("*.mp3"))

records = []
audio_cache = {}

for p in tqdm(audio_paths, desc="Loading audio"):
    speaker = p.parts[-2] if len(p.parts) >= 2 else "unknown"
    audio_id = p.stem
    y, _ = librosa.load(p, sr=16000, mono=True)
    y = y.astype(np.float32)
    duration_sec = len(y) / 16000.0
    audio_cache[str(p)] = y
    records.append({
        "speaker": speaker,
        "audio_path": str(p),
        "audio_id": audio_id,
        "ext": p.suffix.lstrip('.'),
        "duration_sec": duration_sec,
    })

df_audio = pd.DataFrame(records).sort_values(["speaker", "audio_id"]).reset_index(drop=True)
display(df_audio.head())
print("Counts by speaker:", df_audio.groupby("speaker").size())
print("Total files", len(df_audio))

```

## Benchmark Config


```python

import torch, time, gc
from dataclasses import dataclass

beam_size = int(os.getenv("BEAM_SIZE", "5"))
MMS_ENABLED = bool(int(os.getenv("ENABLE_MMS", "0")))

MODELS = [
    {"name": "whisper-tiny", "hf_id": "openai/whisper-tiny", "family": "whisper", "enabled": True},
    {"name": "whisper-base", "hf_id": "openai/whisper-base", "family": "whisper", "enabled": True},
    {"name": "whisper-small", "hf_id": "openai/whisper-small", "family": "whisper", "enabled": True},
    {"name": "whisper-medium", "hf_id": "openai/whisper-medium", "family": "whisper", "enabled": True},
    {"name": "whisper-large-v3", "hf_id": "openai/whisper-large-v3", "family": "whisper", "enabled": True},
    {"name": "seamless-m4t-v2-large", "hf_id": "facebook/seamless-m4t-v2-large", "family": "seamlessm4t", "enabled": True},
    {"name": "mms-1b-all-fa", "hf_id": "facebook/mms-1b-all", "family": "mms_ctc", "language": "fa", "enabled": MMS_ENABLED},
]

BACKENDS = [
    {"name": "transformers_whisper_fp16", "type": "transformers_whisper", "compute_type": "float16", "supported_families": ["whisper"], "enabled": True},
    {"name": "faster_whisper_float16", "type": "faster_whisper", "compute_type": "float16", "supported_families": ["whisper"], "enabled": True},
    {"name": "faster_whisper_int8_float16", "type": "faster_whisper", "compute_type": "int8_float16", "supported_families": ["whisper"], "enabled": True},
    {"name": "transformers_seamless_fp16", "type": "transformers_seamless", "compute_type": "float16", "supported_families": ["seamlessm4t"], "enabled": True},
    {"name": "transformers_mms_ctc_fp16", "type": "transformers_mms_ctc", "compute_type": "float16", "supported_families": ["mms_ctc"], "enabled": MMS_ENABLED},
]

def sync_cuda():
    if torch.cuda.is_available():
        torch.cuda.synchronize()

def safe_empty_cache():
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
    gc.collect()

VALID_COMBOS = []
for m in MODELS:
    if not m.get("enabled", True):
        continue
    for b in BACKENDS:
        if not b.get("enabled", True):
            continue
        if m["family"] in b["supported_families"]:
            VALID_COMBOS.append({
                "model_name": m["name"],
                "model_id": m["hf_id"],
                "family": m["family"],
                "backend": b["name"],
                "backend_type": b["type"],
                "compute_type": b.get("compute_type"),
                "language": m.get("language", "fa"),
            })

df_valid = pd.DataFrame(VALID_COMBOS)
display(df_valid)
print("Total valid combos", len(df_valid))

all_results = []

```

## Run: Transformers Whisper FP16


```python

    import pandas as pd
    from transformers import WhisperProcessor, WhisperForConditionalGeneration

    backend_name = "transformers_whisper_fp16"
    combos = [c for c in VALID_COMBOS if c["backend"] == backend_name]
    backend_results = []

    device = "cuda" if torch.cuda.is_available() else "cpu"
    dtype = torch.float16 if torch.cuda.is_available() else torch.float32

    for combo in combos:
        model_id = combo["model_id"]
        print(f"
Loading {model_id} on {device}")
        try:
            processor = WhisperProcessor.from_pretrained(model_id, local_files_only=LOCAL_FILES_ONLY)
            model = WhisperForConditionalGeneration.from_pretrained(
                model_id,
                torch_dtype=dtype,
                device_map="auto" if torch.cuda.is_available() else None,
                local_files_only=LOCAL_FILES_ONLY,
            )
            model.eval()
        except Exception as e:
            print(f"Load failed for {model_id}: {e}")
            for _, row in df_audio.iterrows():
                backend_results.append({
                    "backend": backend_name,
                    "model_id": model_id,
                    "speaker": row.speaker,
                    "audio_id": row.audio_id,
                    "duration_sec": row.duration_sec,
                    "latency_sec": None,
                    "rtf": None,
                    "pred_text": None,
                    "error": f"load_failed: {e}",
                })
            continue

        try:
            forced_ids = processor.get_decoder_prompt_ids(language="fa", task="transcribe")
        except Exception:
            forced_ids = None

        # warmup on first file
        if len(df_audio):
            y0 = audio_cache[df_audio.iloc[0].audio_path]
            inputs = processor(y0, sampling_rate=16000, return_tensors="pt").input_features.to(device)
            sync_cuda()
            with torch.no_grad():
                _ = model.generate(inputs, forced_decoder_ids=forced_ids, num_beams=beam_size, max_length=448)
            sync_cuda()

        for _, row in df_audio.iterrows():
            y = audio_cache[row.audio_path]
            try:
                inputs = processor(y, sampling_rate=16000, return_tensors="pt").input_features.to(device)
                sync_cuda()
                start = time.perf_counter()
                with torch.no_grad():
                    generated_ids = model.generate(
                        inputs,
                        forced_decoder_ids=forced_ids,
                        num_beams=beam_size,
                        max_length=448,
                    )
                sync_cuda()
                latency = time.perf_counter() - start
                text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]
                rtf = latency / row.duration_sec if row.duration_sec > 0 else None
                backend_results.append({
                    "backend": backend_name,
                    "model_id": model_id,
                    "speaker": row.speaker,
                    "audio_id": row.audio_id,
                    "duration_sec": row.duration_sec,
                    "latency_sec": latency,
                    "rtf": rtf,
                    "pred_text": text,
                    "error": None,
                })
                print(f"{model_id} | {row.audio_id} | latency {latency:.3f}s | text: {text[:80]}")
            except Exception as e:
                backend_results.append({
                    "backend": backend_name,
                    "model_id": model_id,
                    "speaker": row.speaker,
                    "audio_id": row.audio_id,
                    "duration_sec": row.duration_sec,
                    "latency_sec": None,
                    "rtf": None,
                    "pred_text": None,
                    "error": str(e),
                })
                print(f"Error on {row.audio_id}: {e}")
        del model
        safe_empty_cache()

    df_backend = pd.DataFrame(backend_results)
    display(df_backend.head())
    summary = df_backend.groupby("model_id").agg(
        mean_latency=("latency_sec", "mean"),
        mean_rtf=("rtf", "mean"),
        success_rate=("error", lambda x: (x.isna()).mean()),
    ).reset_index()
    display(summary)
    all_results.append(df_backend)

```

```python

    import pandas as pd
    from transformers import AutoProcessor, AutoModelForCTC
    import torch.nn.functional as F

    backend_name = "transformers_mms_ctc_fp16"
    combos = [c for c in VALID_COMBOS if c["backend"] == backend_name]
    backend_results = []
    if not combos:
        print("MMS disabled or no compatible models. Set ENABLE_MMS=1 to enable.")
    device = "cuda" if torch.cuda.is_available() else "cpu"
    dtype = torch.float16 if torch.cuda.is_available() else torch.float32

    for combo in combos:
        model_id = combo["model_id"]
        lang = combo.get("language", "fa")
        print(f"
Loading {model_id} for lang={lang}")
        try:
            processor = AutoProcessor.from_pretrained(model_id, local_files_only=LOCAL_FILES_ONLY)
            model = AutoModelForCTC.from_pretrained(
                model_id,
                torch_dtype=dtype,
                device_map="auto" if torch.cuda.is_available() else None,
                local_files_only=LOCAL_FILES_ONLY,
            )
            model.eval()
            if hasattr(processor, "tokenizer") and hasattr(processor.tokenizer, "set_target_lang"):
                try:
                    processor.tokenizer.set_target_lang(lang)
                except Exception:
                    pass
        except Exception as e:
            print(f"Load failed for {model_id}: {e}")
            for _, row in df_audio.iterrows():
                backend_results.append({
                    "backend": backend_name,
                    "model_id": model_id,
                    "speaker": row.speaker,
                    "audio_id": row.audio_id,
                    "duration_sec": row.duration_sec,
                    "latency_sec": None,
                    "rtf": None,
                    "pred_text": None,
                    "error": f"load_failed: {e}",
                })
            continue

        if len(df_audio):
            y0 = audio_cache[df_audio.iloc[0].audio_path]
            inputs = processor(y0, sampling_rate=16000, return_tensors="pt")
            inputs = {k: v.to(device) for k, v in inputs.items()}
            sync_cuda()
            with torch.no_grad():
                _ = model(**inputs)
            sync_cuda()

        for _, row in df_audio.iterrows():
            audio = audio_cache[row.audio_path]
            try:
                inputs = processor(audio, sampling_rate=16000, return_tensors="pt")
                inputs = {k: v.to(device) for k, v in inputs.items()}
                sync_cuda()
                start = time.perf_counter()
                with torch.no_grad():
                    logits = model(**inputs).logits
                sync_cuda()
                latency = time.perf_counter() - start
                pred_ids = logits.argmax(dim=-1)
                text = processor.batch_decode(pred_ids)[0]
                rtf = latency / row.duration_sec if row.duration_sec > 0 else None
                backend_results.append({
                    "backend": backend_name,
                    "model_id": model_id,
                    "speaker": row.speaker,
                    "audio_id": row.audio_id,
                    "duration_sec": row.duration_sec,
                    "latency_sec": latency,
                    "rtf": rtf,
                    "pred_text": text,
                    "error": None,
                })
                print(f"{model_id} | {row.audio_id} | {latency:.3f}s | text: {text[:80]}")
            except Exception as e:
                backend_results.append({
                    "backend": backend_name,
                    "model_id": model_id,
                    "speaker": row.speaker,
                    "audio_id": row.audio_id,
                    "duration_sec": row.duration_sec,
                    "latency_sec": None,
                    "rtf": None,
                    "pred_text": None,
                    "error": str(e),
                })
                print(f"Error on {row.audio_id}: {e}")

        del model
        safe_empty_cache()

    if backend_results:
        df_backend = pd.DataFrame(backend_results)
        display(df_backend.head())
        summary = df_backend.groupby("model_id").agg(
            mean_latency=("latency_sec", "mean"),
            mean_rtf=("rtf", "mean"),
            success_rate=("error", lambda x: (x.isna()).mean()),
        ).reset_index()
        display(summary)
        all_results.append(df_backend)

```
